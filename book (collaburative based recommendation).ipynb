```python
y = x[x].index
y
```

```python
rating = rating_data[rating_data['User-ID'].isin(y)]
rating
```

```python
ratings_with_book = rating.merge(book_data,on='ISBN')
ratings_with_book
```

```python
num_rating = ratings_with_book.groupby('Book-Title')['Book-Rating'].count().reset_index()
num_rating=num_rating.rename(columns={'Book-Rating':'num_ratings'})
num_rating
```

```python
final_rating = ratings_with_book.merge(num_rating,on='Book-Title')
final_rating
```

```python
final_rating = final_rating[final_rating['num_ratings'] >=50]
final_rating
```

```python
final_rating = final_rating.drop_duplicates(['User-ID','Book-Title'])
final_rating
```

```python
final_rating.shape
```

```python
Pt = final_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
Pt
```

```python
Pt.fillna(0,inplace=True)
```

```python
Pt
```

```python
from scipy.sparse import csr_matrix
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(Pt)
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model = NearestNeighbors()
```

```python
model.fit(book_sparse)
```

```python
distance,suggestion = model.kneighbors(Pt.iloc[237,:].values.reshape(1,-1),n_neighbors=6)
```
