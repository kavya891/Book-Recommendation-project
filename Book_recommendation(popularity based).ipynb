# Popularity Based

 It will suggest items based their overall popularity rather than user performance ,based on metrix like,
   
* Number of purchases (e.g., best-selling books on Amazon)
* Number of views (e.g., trending videos on YouTube)
* Number of ratings (e.g., top-rated restaurants on Google)
* Average rating score (e.g., highest-rated movies on IMDb

---* Popularity based recommendation systems works with the trend.

---* It basically uses the items which are in trend right now.

```python
# merge the data
rating_with_num = rating_data.merge(book_data,on='ISBN')
rating_with_num
```

```python
# number of ratings(we count it)
num_ratings_df = rating_with_num.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_ratings_df
```

```python
# rename the column anme
num_ratings_df = num_ratings_df.rename(columns={'Book-Rating':'num_ratings'})
num_ratings_df
```

```python
# find the average rating
avg_ratings_df = rating_with_num.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_ratings_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_ratings_df
```

```python
print(rating_with_num['Book-Rating'].dtype)
```

```python
rating_with_num['Book-Rating'] = pd.to_numeric(rating_with_num['Book-Rating'], errors='coerce')
```

```python
print(rating_with_num['Book-Rating'].unique().sum())
```

```python
# merge it
popular_df =num_ratings_df.merge(avg_ratings_df,on='Book-Title') 
popular_df
```

```python
# above 250 rating of popular books
popular_df[popular_df['num_ratings']>=250]
```

```python
popular_df[popular_df['num_ratings']>=250].sort_values('avg_rating',ascending=False)
```

```python
popular_df=popular_df[popular_df['num_ratings']>=250].sort_values('avg_rating',ascending=False).head(50)
popular_df
```

# Collabarative based recommnedation system

Collaborative Filtering (CF) is a recommendation technique that suggests items (books, movies, products) based on user interactions. It works by analyzing past behavior and finding patterns between users or items.

1.User-Based Collaborative Filtering

* Finds users with similar preferences and recommends books they liked.
* Example: If Alice and Bob both liked "Harry Potter", and Alice liked "The Hobbit", Bob might like "The Hobbit" too.
* Uses cosine similarity or Pearson correlation between user rating vectors.

2️⃣ Item-Based Collaborative Filtering

* Finds books that have been rated similarly by users.
* Example: If many users who liked "The Hobbit" also liked "Lord of the Rings", they are similar, and we can recommend one if a user reads the other.
* Uses cosine similarity between book rating vectors.
